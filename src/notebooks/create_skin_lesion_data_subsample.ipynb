```python
import pandas as pd
```

```python
DRIVE_DIR = '/content/drive/MyDrive'
PROJECT_ROOT_DIR = f'{DRIVE_DIR}/src'
DATA_DIR = f'{PROJECT_ROOT_DIR}/data'
IMG_DIR = f'{DATA_DIR}/images'
```

```python
FULL_COLUMN_LABELS = {
    "MEL": "Melanoma",
    "NV": "Melanocytic nevus",
    "BCC": "Basal cell carcinoma",
    "AK": "Actinic keratosis",
    "BKL": "Benign keratosis",
    "DF": "Dermatofibroma",
    "VASC": "Vascular lesion",
    "SCC": "Squamous cell carcinoma",
    "UNK": "None of the others"
  }
```

```python
isic_original_data = pd.read_csv(f'{DATA_DIR}/ISIC_2019_Training_GroundTruth.csv')
```

```python
df_temp = isic_original_data.copy()
```

```python
df_mel = df_temp[df_temp['MEL'] == 1.0].sample(n=4500, random_state=42)
df_nv = df_temp[df_temp['NV'] == 1.0].sample(n=845, random_state=42)
df_bcc = df_temp[df_temp['BCC'] == 1.0].sample(n=845, random_state=42)
df_ak = df_temp[df_temp['AK'] == 1.0].sample(n=845, random_state=42)
df_bkl = df_temp[df_temp['BKL'] == 1.0].sample(n=845, random_state=42)
df_df = df_temp[df_temp['DF'] == 1.0]
df_vasc = df_temp[df_temp['VASC'] == 1.0]
df_scc = df_temp[df_temp['SCC'] == 1.0]

# Concatenar todos os DataFrames amostrados
df_subsample = pd.concat([df_mel, df_nv, df_bcc, df_ak, df_bkl, df_df, df_vasc, df_scc]).sample(frac=1)

# Verificar o novo DataFrame
print(df_subsample.head())
print(df_subsample.shape)
```

```python
(((df_subsample.iloc[:, 1:].rename(columns=FULL_COLUMN_LABELS)) == 1.0).sum())
```

```python
df_subsample.to_csv(f'{DATA_DIR}/ISIC_2019_subsample.csv', sep=',', index=False, encoding='utf-8')
```

```python
df_balanced = df_subsample[['image', 'MEL']].copy().rename(columns={'MEL': 'label'})
```

```python
df_balanced.info()
```

```python
df_balanced.describe()
```

```python
labels = df_balanced['label'].unique().tolist()
labels.sort()
print(labels)
```

```python
melanoma_samples = (((df_balanced == 1.0).values).sum())
non_melanoma_samples = (((df_balanced == 0.0).values).sum())

print('Dataframe samples balanced')
print('------------------')
print('Melanoma:', melanoma_samples)
print('Non-melanoma:', non_melanoma_samples)
```

```python
import sklearn
sklearn.__version__
```

```python
from sklearn.model_selection import train_test_split
```

```python
# split the data into train and temp set
train_data, temp_data = train_test_split(df_balanced, test_size=0.3, random_state=42, shuffle=True, stratify=df_balanced['label'])
```

```python
train_data.reset_index(drop=True, inplace=True)
train_data.head(3)
```

```python
train_non_melanoma_samples = (((train_data == 0.0).values).sum())
train_melanoma_samples = (((train_data == 1.0).values).sum())

print('Train samples')
print('------------------')
print('Non-melanoma:', train_non_melanoma_samples)
print('Melanoma:', train_melanoma_samples)
print('------------------')
print('Total:', train_melanoma_samples + train_non_melanoma_samples)
```

```python
temp_non_melanoma_samples = (((temp_data == 0.0).values).sum())
temp_melanoma_samples = (((temp_data == 1.0).values).sum())

print('Temp data samples')
print('------------------')
print('Non-melanoma:', temp_non_melanoma_samples)
print('Melanoma:', temp_melanoma_samples)
print('------------------')
print('Total:', temp_melanoma_samples + temp_non_melanoma_samples)
```

```python
# split the temp data into validate and test set
val_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42, shuffle=True, stratify=temp_data['label'])
```

```python
val_data.reset_index(drop=True, inplace=True)
val_data.head(3)
```

```python
val_non_melanoma_samples = (((val_data == 0.0).values).sum())
val_melanoma_samples = (((val_data == 1.0).values).sum())

print('Validate samples')
print('------------------')
print('Non-melanoma:', val_non_melanoma_samples)
print('Melanoma:', val_melanoma_samples)
print('------------------')
print('Total:', val_melanoma_samples + val_non_melanoma_samples)
```

```python
test_data.reset_index(drop=True, inplace=True)
test_data.head(3)
```

```python
test_non_melanoma_samples = (((test_data == 0.0).values).sum())
test_melanoma_samples = (((test_data == 1.0).values).sum())

print('Test samples')
print('------------------')
print('Non-melanoma:', test_non_melanoma_samples)
print('Melanoma:', test_melanoma_samples)
print('------------------')
print('Total:', test_melanoma_samples + test_non_melanoma_samples)
```
